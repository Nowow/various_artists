```python
from collections import Counter, defaultdict

class LanguageModel:
    def __init__(self, data, order=4):
        self.order = order
        self.ngrams = defaultdict(Counter)
        pad = '~' * order
        data = pad + data
     
        # For 1 symbol
        for i in range (0, len(data)-order):
            self.ngrams[data[i:i+order]][data[i+order]] += 1

        self.lm = {history: self.normalize(chars) for history, chars in self.ngrams.items()}
    
    def normalize(self, counter):
        ### YOUR CODE HERE
        # Normalize entries of counter.
        normValues = []
        for i in counter:
            normValues.append((i, counter[i]/sum(counter.values())))
        return normValues
        ### END YOUR CODE
    
    def __getitem__(self, history):
        return self.lm[history]
```

```python
from numpy.random import choice

def generate_letter(lm, history):
    history = history[-lm.order:]
    ### YOUR CODE HERE
    # Generate the next character according to the history.
    # Sample the next letter according to your probability distribution.
    
    choices = [i[0] for i in lm[history]]
    pDist = [i[1] for i in lm[history]]
    decision = choice(choices, 1, pDist)[0]

    return decision
    ### END YOUR CODE
        
def generate_text(lm, n_letters=1000):
    history = '~' * lm.order
    out = []
    ### YOUR CODE HERE
    # Generate random text and stash it into out variable.
    for i in range(n_letters):
        symbol = generate_letter(lm, history)
        history += symbol
        out.append(symbol)

    ### END YOUR CODE
    return ''.join(out)
```
